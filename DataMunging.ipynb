```python
import pandas as pd
import sqlalchemy as sq
import numpy as np
pd.set_option("display.max_rows",200)
pd.set_option("display.max_columns",999)
import warnings
warnings.filterwarnings('ignore')
```

```python
engine = sq.create_engine('sqlite:///OKCdatabase.db') #create connection to the database
```

# User attributes data table

```python
df_u = pd.read_sql_table('user_info',engine)
df_u = df_u.drop('d_languages',1)
df_u = df_u.drop('d_username',1)
df_u = df_u.drop('d_city',1)
df_u = df_u.drop('d_astrology_sign',1)
df_u = df_u.drop('d_country',1)
#set aside age and income, as they are numerical
age = df_u.d_age
df_u.drop('d_age', axis=1,inplace=True)
income = df_u.d_income
df_u.drop('d_income', axis=1,inplace=True)
```

```python
# clean up any spaces, turn them into underscores
def spaces(item):
    if item:
        if isinstance(item, str):
            return item.replace(" ","_")
        else:
            return item
    else:
        return item
    
```

```python
# underscore df_U
df_u = df_u.applymap(spaces)
```

```python
# clean up and dummy everything within df_U

df_u['d_gender_Man'] = df_u['d_gender'].astype(str).str.contains('((?<!Trans_)(?<!Cis_)Man)',regex=True).astype(int)
df_u['d_gender_Woman'] = df_u['d_gender'].astype(str).str.contains('((?<!Trans_)(?<!Cis_)Woman)',regex=True).astype(int)
df_u['d_gender_Cis_Man'] = df_u['d_gender'].astype(str).str.contains('(_*Cis_Man)',regex=True).astype(int)
df_u['d_gender_Cis_Womanan'] = df_u['d_gender'].astype(str).str.contains('(_*Cis_Woman)',regex=True).astype(int)
df_u['d_gender_Agender'] = df_u['d_gender'].astype(str).str.contains('(_*Agender)',regex=True).astype(int)
df_u['d_gender_Androgynous'] = df_u['d_gender'].astype(str).str.contains('(_*Androgynous)',regex=True).astype(int)
df_u['d_gender_Bigender'] = df_u['d_gender'].astype(str).str.contains('(_*Bigender)',regex=True).astype(int)
df_u['d_gender_Genderfluid'] = df_u['d_gender'].astype(str).str.contains('(_*Genderfluid)',regex=True).astype(int)
df_u['d_gender_Genderqueer'] = df_u['d_gender'].astype(str).str.contains('(_*Genderqueer)',regex=True).astype(int)
df_u['d_gender_Intersex'] = df_u['d_gender'].astype(str).str.contains('(_*Intersex)',regex=True).astype(int)
df_u['d_gender_Non_binary'] = df_u['d_gender'].astype(str).str.contains('(_*Non-binary)',regex=True).astype(int)
df_u['d_gender_Other'] = df_u['d_gender'].astype(str).str.contains('(_*Other)',regex=True).astype(int)
df_u['d_gender_Trans_Man'] = df_u['d_gender'].astype(str).str.contains('(_*Trans_man)',regex=True).astype(int)
df_u['d_gender_Trans_Woman'] = df_u['d_gender'].astype(str).str.contains('(_*Trans_Woman)',regex=True).astype(int)
df_u['d_gender_Transfeminine'] = df_u['d_gender'].astype(str).str.contains('(_*Transfeminine)',regex=True).astype(int)
df_u['d_gender_Transmasculine'] = df_u['d_gender'].astype(str).str.contains('(_*Transmasculine)',regex=True).astype(int)
df_u['d_gender_Transmasculine'] = df_u['d_gender'].astype(str).str.contains('(_*Transgender)',regex=True).astype(int)
df_u = df_u.drop('d_gender', axis=1)
df_u['d_Asian'] = df_u['d_ethnicity'].astype(str).str.contains('(Asian)',regex=True).astype(int)
df_u['d_Black'] = df_u['d_ethnicity'].astype(str).str.contains('(Black)',regex=True).astype(int)
df_u['d_Native_American'] = df_u['d_ethnicity'].astype(str).str.contains('(Native American)',regex=True).astype(int)
df_u['d_Pacific_Islander'] = df_u['d_ethnicity'].astype(str).str.contains('(Pacific Islander)',regex=True).astype(int)
df_u['d_Hispanic'] = df_u['d_ethnicity'].astype(str).str.contains('(Hispanic / Latin)',regex=True).astype(int)
df_u['d_Indian'] = df_u['d_ethnicity'].astype(str).str.contains('(Indian)',regex=True).astype(int)
df_u['d_White'] = df_u['d_ethnicity'].astype(str).str.contains('(White)',regex=True).astype(int)
df_u['d_Middle_Eastern'] = df_u['d_ethnicity'].astype(str).str.contains('(Middle Eastern)',regex=True).astype(int)
df_u['d_Other'] = df_u['d_ethnicity'].astype(str).str.contains('(Other)',regex=True).astype(int)
df_u.drop('d_ethnicity', axis=1,inplace=True)
df_u['d_Asexual'] = df_u['d_orientation'].astype(str).str.contains('(Asexual)',regex=True).astype(int)
df_u['d_Queer'] = df_u['d_orientation'].astype(str).str.contains('(Queer)',regex=True).astype(int)
df_u['d_Questioning'] = df_u['d_orientation'].astype(str).str.contains('(Questioning)',regex=True).astype(int)
df_u['d_Straight'] = df_u['d_orientation'].astype(str).str.contains('(Straight)',regex=True).astype(int)
df_u['d_Heteroflexible'] = df_u['d_orientation'].astype(str).str.contains('(Heteroflexible)',regex=True).astype(int)
df_u['d_Sapiosexual'] = df_u['d_orientation'].astype(str).str.contains('(Sapiosexual)',regex=True).astype(int)
df_u['d_Demisexual'] = df_u['d_orientation'].astype(str).str.contains('(Demisexual)',regex=True).astype(int)
df_u['d_Bisexual'] = df_u['d_orientation'].astype(str).str.contains('(Bisexual)',regex=True).astype(int)
df_u['d_Lesbian'] = df_u['d_orientation'].astype(str).str.contains('(Lesbian)',regex=True).astype(int)
df_u['d_Homoflexible'] = df_u['d_orientation'].astype(str).str.contains('(Homoflexible)',regex=True).astype(int)
df_u['d_Pansexual'] = df_u['d_orientation'].astype(str).str.contains('(Pansexual)',regex=True).astype(int)
df_u['d_Gay'] = df_u['d_orientation'].astype(str).str.contains('(Gay)',regex=True).astype(int)
df_u.drop('d_orientation', axis=1,inplace=True)
df_u.d_religion_type.replace('-',np.nan,inplace=True)
df_u.d_religion_type.fillna(np.nan,inplace=True)
df_u_bin = pd.get_dummies(df_u)
df_u_bin.head()
```

# Match preferences

```python
df_m = pd.read_sql_table('match_preferences',engine)

df_m = df_m.applymap(spaces)

df_m.fillna(value=np.nan,inplace=True)
df_m['lf_for_new_friends'] = df_m['lf_for'].astype(str).str.contains('(new friends)',regex=True).astype(int)
df_m['lf_for_long_term_dating'] = df_m['lf_for'].astype(str).str.contains('(long-term dating)',regex=True).astype(int)
df_m['lf_for_short_term_dating'] = df_m['lf_for'].astype(str).str.contains('(short-term dating)',regex=True).astype(int)
df_m['lf_for_casual_sex'] = df_m['lf_for'].astype(str).str.contains('(casual sex)',regex=True).astype(int)
df_m.drop('lf_for', axis=1,inplace=True)
lf_rest = pd.get_dummies(df_m.loc[:,['lf_single','lf_want','lf_location']])
df_m = pd.concat([df_m, lf_rest],axis=1)
df_m.drop('lf_single', axis=1,inplace=True)
df_m.drop('lf_want', axis=1,inplace=True)
df_m.drop('lf_location', axis=1,inplace=True)
#set aside age preferences, as they are numerical
max_age = df_m.lf_max_age
min_age = df_m.lf_min_age
df_m.drop('lf_max_age', axis=1,inplace=True)
df_m.drop('lf_min_age', axis=1,inplace=True)
df_m_bin = df_m
df_m_bin.head()
```

# Question responses

```python
df_q = pd.read_sql_table('question_responses1',engine);
```

```python
#dataframe of general questions
df_qs = pd.read_csv('question_data.csv',sep=';',index_col=None)
df_qs = df_qs.applymap(spaces)
df_qs = df_qs.sort_values('N',ascending=False);
#dataframe of cognitive test questions
df_ts = pd.read_csv('test_items.csv',index_col=None)
#list of all of the question headers (sorted by number of responses):
headers_all = list(df_qs.iloc[:,0])
#list of headers that starts with q
headers_q = [name for name in headers_all if (name[0] == 'q')]
#list of headers that starts with If
headers_lf = [name for name in headers_all if (name[0] == 'l')]
#list of headers that starts with d
headers_d = [name for name in headers_all if (name[0] == 'd')]
#list of headers that starts with p
headers_p = [name for name in headers_all if (name[0] == 'p')]
#list of cognitive test question headers
headers_ts = list(df_ts.iloc[:,0])
#list of non-cognitive test question headers
headers_qns = [name for name in headers_q if (name not in headers_ts)]
header_O = list(df_qs[df_qs['Type']=='O'].iloc[:,0])
header_M = list(df_qs[df_qs['Type']=='M'].iloc[:,0])
header_N = list(df_qs[df_qs['Type']=='N'].iloc[:,0])
headers_O = [name for name in headers_qns[0:200] if (name in header_O)]
df_qO = df_q.loc[:,headers_O]
headers_N = [name for name in headers_qns[0:200] if (name in header_N)]
df_qN = df_q.loc[:,headers_N]
headers_M = [name for name in headers_qns[0:200] if (name in header_M)]
df_qM = df_q.loc[:,headers_M]
```

#### now we need to remove spaces on all question responses prior to get_dummies


Ordinal Responses

```python
# de-space and dummy ordinal responses
df_qO = df_qO.applymap(spaces)
df_qO_bin = pd.get_dummies(df_qO);df_qO_bin.head()
```

Nominal Responses

```python
# de-space and dummy nominal responses
df_qN= df_qN.applymap(spaces)
df_qN_bin = pd.get_dummies(df_qN);df_qN_bin.head()
```

Mixed Responses

```python
# de-space and dummy mixed responses
df_qM = df_qM.applymap(spaces)
df_qM_bin = pd.get_dummies(df_qM);df_qM_bin.head()
```

## Write a new database with the processed tables for modelling purposes

```python
engine2 = sq.create_engine('sqlite:///OKC_Processed_DB.db') #create connection to the database
```

### Numerical Table

```python
# make income numeric
income.replace('$40,000-$50,000','45000',inplace=True)
income.replace('$70,000-$80,000','75000',inplace=True)
income.replace('$20,000-$30,000','25000',inplace=True)
income.replace('Less than $20,000','10000',inplace=True)
income.replace('$100,000-$150,000','125000',inplace=True)
income.replace('$250,000-$500,000','375000',inplace=True)
income.replace('$30,000-$40,000','35000',inplace=True)
income.replace('$500,000-$1,000,000','750000',inplace=True)
income.replace('$80,000-$100,000','90000',inplace=True)
income.replace('$50,000-$60,000','55000',inplace=True)
income.replace('More than $1,000,000','1000000',inplace=True)
income.replace('Rather not say',np.nan,inplace=True)
income.fillna(np.nan,inplace=True)
```

```python
#these are the tables with numerical entries
tab_num = pd.concat([income,age,max_age,min_age,df_p,df_CA],axis=1)
tab_num.head()
```

```python
tab_num.to_sql('numerical_attributes',engine2,index=False)
```
